```python
import pandas as pd
df = pd.read_csv('xTest.csv')
```

```python
df = df.drop(columns=['Rk','Awards','Player-additional','Pos'])
df.head()
```

```python
numRowDrop = 0
indtoDrop = []
for index, row in df.iterrows():
    if numRowDrop>0:
        indtoDrop.append(index)
        numRowDrop-=1
    elif str(row[2]).endswith('TM'):
        numRowDrop = int(row[2][0])
df = df.drop(index=indtoDrop).reset_index(drop=True)

df.to_csv('xTest.csv', index=False)
```
